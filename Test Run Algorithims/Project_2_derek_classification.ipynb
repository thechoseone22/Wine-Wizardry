```python
#creating target variable column for wine type classiffication model
red_df['wine_type'] = 'red'
white_df['wine_type'] = 'white'
```

```python
red_df
```

```python
white_df
```

```python
#concatenating datasets and classifying the red and white wine into 0 & 1
combined_df = pd.concat([red_df, white_df], ignore_index=True)
combined_df['wine_type'] = combined_df['wine_type'].apply(lambda x: 0 if x == 'red' else 1)
```

```python
combined_df
```

```python
#checking for missing data again
combined_df.isnull().sum()
```

```python
#creating our features and target variables
X = combined_df.drop(['wine_type', 'quality'], axis=1)
y = combined_df['wine_type']
```

```python
#checking for imbalanced data
combined_df['wine_type'].value_counts()
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
#using smote to balance data
smote=SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
#checking for balanced data
y_train_smote.value_counts()
```

```python
#fitting train test split randomforestclassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_smote, y_train_smote)
```

```python
y_pred = model.predict(X_test)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
grid_search.fit(X_train, y_train)

best_rf_model = grid_search.best_estimator_
```

```python
y_pred = best_rf_model.predict(X_test)
```

```python
y_pred
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
#classification report with accuracy as well
report = classification_report(y_test, y_pred, target_names=['Red', 'White'])
print("Classification Report:")
print(report)
```

```python
#showing feature importances. which shows which features correlated the most with predicting our target variable (wine type)
importances = best_rf_model.feature_importances_
indices = np.argsort(importances)[::-1]
feature_names = X.columns

plt.figure()
plt.title("Feature Importance")
plt.bar(range(X.shape[1]), importances[indices], align="center")
plt.xticks(range(X.shape[1]), feature_names[indices], rotation=90)
plt.xlim([-1, X.shape[1]])
plt.show()
```

```python
feature_importances = best_rf_model.feature_importances_
features = X.columns

plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances, y=features)
plt.title('Which feature had most impact in classifying red vs white wine')
plt.show()
```

```python
#confusion matrix showing how accurate our model is visually
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False, xticklabels=['Red', 'White'], yticklabels=['Red', 'White'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
